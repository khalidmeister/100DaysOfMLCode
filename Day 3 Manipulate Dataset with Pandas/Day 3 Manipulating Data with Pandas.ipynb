## Differences between [] and [[]]

Sebenarnya perbedaannya adalah dari tipe object yang diambil. Hal ini penting karena kalau mau melakukan proses dengan NumPy biasanya digunakan object Series. 

Kalau memilih kolom pakai [] akan mengeluarkan tipe Series, sedangkan [[]] mengeluarkan tipe DataFrame

```python
series = df['Medal']
print(series.head())
print(type(series))
```

```python
series = df[['Medal']]
print(series.head())
print(type(series))
```

## Filtering the Dataframe

Misalnya ingin mencari tahu siapa peraih medali emas cabor renang di Olimpiade 2008 di Beijing. Maka dilakukan filtering bertahap dari kolom edition, discipline, lalu medal

```python
df_beijing = df.loc[df['Edition'] == 2008, :]
```

```python
df_beijing.head()
```

```python
df_beijing_swimming = df_beijing.loc[df['Discipline'] == 'Swimming', :]
```

```python
df_beijing_swimming_gold = df_beijing_swimming.loc[df['Medal'] == 'Gold', :]
```
